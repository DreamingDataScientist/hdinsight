# Pyspark - StreamingContext

```pyspark
from pyspark.streaming import StreamingContext
```

```pyspark
ssc = StreamingContext(sc, 3)
```

```pyspark
ssc.checkpoint("\chkpnt")
```

```pyspark
stream_txt = ssc.textFileStream("\stream")
```

```pyspark
words = stream_txt.flatMap(lambda line: line.split(" "))
pairs = words.map(lambda word: (word,1))
word_count = pairs.reduceByKeyAndWindow(lambda a, b: a +b, lambda x,y : x-y, 60, 3)
sorted_word_count = word_count.transform(lambda rdd: rdd.sortBy(lambda x: x[1], ascending=False))
```

```pyspark
sorted_word_count.pprint(num=20)
```

```pyspark
ssc.start()
```

```pyspark
ssc.stop()
```
